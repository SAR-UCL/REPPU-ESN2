```python
import pandas as pd
import numpy as np

# Number of days
num_days = 20
# Number of 5-minute intervals per day
num_intervals_per_day = 288
# Total number of intervals
total_intervals = num_days * num_intervals_per_day

# Generate datetime values with 5-minute intervals
start_date = pd.Timestamp('2024-01-01')
date_range = pd.date_range(start=start_date, periods=total_intervals, freq='5min')

# Generate random data
data = np.random.randn(total_intervals)

# Create DataFrame
df = pd.DataFrame({'Date': date_range,
                   'Value': data})

# Shuffle the dates
shuffled_dates = np.random.permutation(pd.unique(df['Date'].dt.date))

# Split into 18 days and 2 days
train_dates = shuffled_dates[:18]
test_dates = shuffled_dates[18:]

# Filter DataFrame based on the selected dates
train_df = df[df['Date'].dt.date.isin(train_dates)]
test_df = df[df['Date'].dt.date.isin(test_dates)]

train_df
```

```python
import numpy as np
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([1, 2, 4, 4, 5])

err_pct = np.mean(np.abs(arr2 - arr1) / arr1) * 100
mae = np.mean(np.abs(arr2 - arr1))
mse = np.mean((arr2 - arr1) ** 2)
rmse = np.sqrt(mse)

print(f"Error Percentage: {err_pct:.2f}%")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
import numpy as np

# Generate some sample data
y_true = np.array([1, 2, 3, 4, 5])
y_pred = np.array([1.5, 2.5, 2.8, 4.2, 4.9])

y_true = np.array([1, 2, 3, 4, 5])
y_pred = np.array([1, 2, 4, 4, 5])

# Calculate the MAE
mae = np.mean(np.abs(y_true - y_pred))
mape = np.mean(np.abs((y_pred - y_true) / y_true)) * 100
rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))

print("Mean Absolute Error:", mae)
print("Mean Absolute Percentage Error:", mape)
print("Root Mean Squared Error:", rmse)
```
