```python
#write a fibonacci sequence generator with 30 values then plot

def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fib = [fibonacci(i) for i in range(30)]

plt.plot(fib)
```
