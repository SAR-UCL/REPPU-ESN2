```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def interest_calc(initial, montly_input, annual_rate, years):
    total = initial
    for i in range(years):
        total = total*(1+annual_rate) + montly_input*12

    return total

interest_calc(1000, 600, 0.08, 3)
```

```python
#a is random array with afixed seed
np.random.seed(0)
a = np.random.rand(20)
b = a - 1
c = a
d = a + np.random.normal(0, 0.4, a.size)
#d = 0.5 * np.sin(2 * np.pi * a) + np.random.normal(0, 0.1, a.size)

R_ab = np.corrcoef(a, b)[0, 1]
RMSE_ab = np.sqrt(np.mean((a - b)**2))

R_ac = np.corrcoef(a, c)[0, 1]
RMSE_ac = np.sqrt(np.mean((a - c)**2))

R_ad = np.corrcoef(a, d)[0, 1]
RMSE_ad = np.sqrt(np.mean((a - d)**2))

figs, ax = plt.subplots(1, 3, figsize=(9, 3))

x = np.linspace(0, 1, a.size)
ax[0].plot(x, a, color='r', label='a')
ax[0].plot(x, b, color='b', label='b')
ax[0].set_title(f'R = {R_ab}\nRMSE = {RMSE_ab}')
ax[0].legend()

ax[1].plot(x, a, color='r', label='a')
ax[1].plot(x, c, color='g', label='c')
ax[1].set_title(f'R = {R_ac}\nRMSE = {RMSE_ac}')
ax[1].legend()

ax[2].plot(x, a, color='r', label='a')
ax[2].plot(x, d, color='m', label='d')
ax[2].set_title(f'R = {R_ad:.2f}\nRMSE = {RMSE_ad:.2f}')
ax[2].legend()

for axs in ax:
    axs.spines['top'].set_visible(False)
    axs.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('errors_vs_correlation_example.png', dpi=300)
```
